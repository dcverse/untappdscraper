```python
def breweryextract(brewery_input):
    #Function takes string name dictionary key which should be tied to Untappd venue url
    brewery_url_input = brewery_list[brewery_input]
    url = brewery_url_input
    beautiful = urllib.request.urlopen(url).read()

    soup = BeautifulSoup(beautiful, 'html.parser')

    #Isolates HTML to relevant beer names lines
    beer_name = soup.find_all('a', {'data-href': ':beer', 'data-track': 'menu'})

    #Parses beer names lines to extract names
    pattern = ">(.*?)\</a>"
    beer_name_list = []
    for i in beer_name:
        i = str(i)
        substring = re.search(pattern, i).group(1)
        beer_name_list.append(substring)

    #Isolates HTML to relevant beer style lines
    beer_style = soup.find_all('em')

    #Parses beer style lines to extract styles
    pattern = "<em>(.*?)\</em>"
    beer_style_list = []
    for i in beer_style:
        i = str(i)
        substring = re.search(pattern, i).group(1)
        beer_style_list.append(substring)

    #Isolates HTML to relevant beer abv and rating lines
    beer_abv_rtg = soup.find_all('h6')

    #Parses beer abv and rating lines to extract ABVs, additional try/except/pass so will not fail if no ABV
    pattern = "<h6><span>(.*?)\ "
    beer_abv_list = []
    for i in beer_abv_rtg:
        try: 
            i = str(i)
            substring = re.search(pattern, i).group(1)
            substring = substring.strip('\%')
            substring = float(substring)
        except:
            substring = None
            pass
        beer_abv_list.append(substring)

    #Parses beer abv and rating lines to extract ratings, additional try/except/pass so will not fail if no rating
    pattern = "data-rating=(.*?)\>"
    beer_rtg_list = []
    for i in beer_abv_rtg:
        try: 
            i = str(i)
            substring = re.search(pattern, i).group(1)
            substring = substring.strip('\"')
            substring = float(substring)
        except:
            substring = None
            pass
        beer_rtg_list.append(substring)

    #Creates list with brewery name to attach to rest of data
    beer_brew_list = []
    for i in beer_abv_rtg:
        substring = brewery_input
        beer_brew_list.append(substring)

    #Creates dictionary of all previously extracted info
    beer_list = {'Brewery': beer_brew_list,
                 'Beer Name': beer_name_list,
                 'Style': beer_style_list,
                 'ABV': beer_abv_list,
                 'Rating': beer_rtg_list}
    return beer_list
    #Convert the dictionary into DataFrame
    #df = pd.DataFrame(beer_list)
    #return df
```
